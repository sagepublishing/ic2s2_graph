````python
neo_graph = open("neo4j_graph.json").read()
neo_graph_data = json.loads(neo_graph)
print(neo_graph_data["results"][0]["data"][3])
````

````python
unique_nodes = [] 
nodes = []
node_ids = []
links = [] 
for connection in neo_graph_data["results"][0]["data"]:
    for node in connection["graph"]["nodes"]:
        if node not in unique_nodes:
            unique_nodes.append(node)
            name = node["properties"]["name"]
            node_type = node["labels"][0]
            node_id = node["id"]
            node_ids.append(node_id)
            node_output = {"id": node_id, "name": name, "type": node_type}
            nodes.append(node_output)
    link_ids = (node_ids.index(connection["graph"]["relationships"][0]['startNode']),connection["graph"]["relationships"][0]['endNode'])
    link = {"source":int(link_ids[0]), "target":int(link_ids[1]), "weight":1}
    links.append(link)
d3_graph = {"nodes": nodes, "links": links}
print (json.dumps(d3_graph))
````
